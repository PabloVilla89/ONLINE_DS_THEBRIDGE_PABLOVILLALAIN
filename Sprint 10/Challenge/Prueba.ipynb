```python
import pandas as pd
import seaborn as sns
from toolbox_ML import (describe_df, tipifica_variables,get_features_num_regression, plot_features_num_regression,get_features_cat_regression,plot_features_cat_regression,plot_features_cat_regression2)
from sklearn.datasets import fetch_openml

# Cargar ejemplo de dataset 
df = sns.load_dataset("penguins")
# 1) Describir el dataset
desc = describe_df(df)
print(desc)

# 2) Tipificar las variables
tipos_sugeridos = tipifica_variables(df, umbral_categoria=10, umbral_continua=0.2)
print(tipos_sugeridos)

target='body_mass_g'
# 3) Seleccionar features numéricas para regresión 
features_num = get_features_num_regression(df, target, umbral_corr=0.2, pvalue=0.05, tipo_variables=tipos_sugeridos)
print("Features numéricas candidatas:", features_num)

# 4) Pairplots de las features seleccionadas
plot_features = plot_features_num_regression(df, target_col=target, columns=features_num, 
                                             umbral_corr=0.2, pvalue=0.05)
print("Features finalmente graficadas:", plot_features)

# 5) Seleccionar features categóricas para regresión 
features_cat = get_features_cat_regression(df, target, pvalue=0.05)
print("Features categóricas candidatas:", features_cat)

# 6) Plots de las variables categóricas significativas
plot_cat_features = plot_features_cat_regression(df, target_col = target, columns = features_cat, p_value = 0.05, with_individual_plot = False)
print("Features categóricas finalmente graficadas:", plot_cat_features)

7
plot_cat_features2 = plot_features_cat_regression2(df, target_col = target, columns = features_cat, pvalue = 0.05, with_individual_plot = False)
print("Features categóricas finalmente graficadas2:", plot_cat_features)
```
